```python
import turtle

# Cấp độ 1: Màu nền cho nhà và ông mặt trời


# Cấp độ 1: Vẽ nhà 1 tầng
turtle.penup()
turtle.goto(-100, -100)
turtle.pendown()
turtle.forward(200)
turtle.left(90)
turtle.forward(200)
turtle.left(90)
turtle.forward(200)
turtle.left(90)
turtle.forward(200)

# Cấp độ 1: Vẽ ông mặt trời
turtle.penup()
turtle.goto(100, 100)
turtle.pendown()
turtle.circle(30)

# Cấp độ 2: Thay đổi màu nền cho cửa sổ và cửa ra vào

# Cấp độ 2: Thêm cửa sổ và cửa ra vào
turtle.penup()
turtle.goto(-50, 50)
turtle.pendown()
turtle.forward(50)
turtle.left(90)
turtle.forward(50)
turtle.left(90)
turtle.forward(50)
turtle.left(90)
turtle.forward(50)

turtle.penup()
turtle.goto(-150, -100)
turtle.pendown()
turtle.forward(50)
turtle.left(90)
turtle.forward(20)
turtle.left(90)
turtle.forward(50)
turtle.left(90)
turtle.forward(20)

# Cấp độ 3: Thay đổi màu nền cho cây

# Cấp độ 3: Thêm cây
turtle.penup()
turtle.goto(150, -100)
turtle.pendown()
turtle.forward(20)
turtle.left(90)
turtle.forward(40)
turtle.right(90)
turtle.forward(30)
turtle.right(90)
turtle.forward(20)
turtle.left(90)
turtle.forward(20)
turtle.right(90)
turtle.forward(10)
turtle.left(90)
turtle.forward(10)

turtle.done()
```
